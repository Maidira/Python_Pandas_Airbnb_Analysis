### Top 5 Hosts by Listing Count and Their Average Rating

```python
df["name"]
```

```python
# Extracting rating from name
df['rating'] = df['name'].str.extract(r'★([0-9.]+)')
print(df[['name', 'rating']])
```

```python
# Average Rating based on host ID
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')

avg_rating_by_host = df.groupby('host_id')['rating'].mean().reset_index()
print(avg_rating_by_host)
```

```python
# Top 5 hosts as per rating
top_5_hosts = avg_rating_by_host.sort_values(by='rating', ascending=False).head(5)
print(top_5_hosts)
```

```python
# Top 5 hosts based on rating and number of reviews they received
host_listings_count = df.groupby('host_id').size().reset_index(name='listing_count')

host_data = pd.merge(avg_rating_by_host, host_listings_count, on='host_id')

top_5_hosts_by_count = host_data.sort_values(by='listing_count', ascending=False).head(5)
print(top_5_hosts_by_count)
```

```python
hosts = top_5_hosts_by_count['host_id'].astype(str)  # Converting to string for better readability
listing_counts = top_5_hosts_by_count['listing_count']
average_ratings = top_5_hosts_by_count['rating']

fig, ax1 = plt.subplots(figsize=(10, 6))

# Bar chart for listing count
ax1.bar(hosts, listing_counts, color='skyblue', label='Listing Count')
ax1.set_xlabel('Host ID')
ax1.set_ylabel('Number of Listings', color='skyblue')
ax1.tick_params(axis='y', labelcolor='skyblue')

# Create a secondary axis for the average rating
ax2 = ax1.twinx()
ax2.plot(hosts, average_ratings, color='orange', marker='o', label='Average Rating')
ax2.set_ylabel('Average Rating', color='orange')
ax2.tick_params(axis='y', labelcolor='orange')


plt.title('Top 5 Hosts by Listing Count and Their Average Rating')
plt.show()
```

### Average avilability by Host
